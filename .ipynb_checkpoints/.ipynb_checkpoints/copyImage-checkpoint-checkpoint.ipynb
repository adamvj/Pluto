```python
import numpy as np
import cv2
import pandas as pd


data = pd.read_csv('./data2/plutoMapping2.csv')     # reading the csv file
data.head()
X = [ ]     # creating an empty array
for img_name in data.IMG_ID:
    img = cv2.imread('' + img_name,1)
    h, w, c = img.shape
    print(h)
    if h > 1280:
        img_half = cv2.resize(img, (0, 0), fx=0.3, fy=0.3)
        h, w, c = img_half.shape
        print(h)
        filename = img_name.replace('data2','data3')
        filename = filename.replace('Pluto2','Pluto')  
        print(filename)
        cv2.imwrite(filename, img_half)
print('Done')
    
```
